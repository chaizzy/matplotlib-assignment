```python
# ANS 2:
# "scatter plot" is type of plot in matplotlib library  where data  is represented  as individual points in 2D 
# according to  given axis 
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel(" x axis")
plt.ylabel(" y axis")
plt.title(" scatter plot")
plt.scatter(x,y)
plt.show()
```

```python
# ANS 3:

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,1,1)
plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
```

```python
# ANS 5:
# bar plot is type of plot which represent the categorical data interms of bar like structure
# bar represent a particular type of data in graph
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

```python
# ANS 6:
# "BOX PLOT" is a type of a plot which describes the numerical data 
# it distributes in five number summary min ,q1 , mid , q3 , max 
# it is used for finding outliers
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1,box2]
plt.boxplot(data)
```
